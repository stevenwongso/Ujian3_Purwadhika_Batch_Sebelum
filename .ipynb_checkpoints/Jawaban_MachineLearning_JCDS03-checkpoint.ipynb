### 1.1

```python
dfJPI = pd.read_excel('indo_12_1.xls',skiprows=3)
dfJPI.rename(columns = {'Unnamed: 0' : 'Provinsi'}, inplace = True)
dfJPI.drop(labels=[34,35], inplace=True)
dfJPI
```

```python
dfJPInew = dfJPI.iloc[[11,6,33]].copy()
dfJPInew2 = dfJPInew.transpose()
dfJPInew2.drop('Provinsi',inplace=True)
dfJPInew2.columns = ['JB','B','I']
dfJPInew2['Tahun']= dfJPInew2.index
dfJPInew2
```

```python
plt.style.available
plt.style.use('seaborn')
```

```python
# plt.style.use('Seaborn')
plt.figure(figsize=(12,6))
plt.plot(dfJPInew2['Tahun'],dfJPInew2['JB'],'g-o')
plt.plot(dfJPInew2['Tahun'],dfJPInew2['B'],'b-o')
plt.plot(dfJPInew2['Tahun'],dfJPInew2['I'],'r-o')

plt.xlabel('Tahun'); plt.ylabel('jumlah penduduk(ratus juta jiwa)')
plt.yticks(np.arange(0,300000000,50000000))
plt.legend(dfJPInew['Provinsi'])
plt.show()
```

<hr>

### 1.2

```python
from sklearn.linear_model import LinearRegression
```

```python
modelJB = LinearRegression()
modelB = LinearRegression()
modelI = LinearRegression()

modelJB.fit(dfJPInew2[['Tahun']],dfJPInew2['JB'])
modelB.fit(dfJPInew2[['Tahun']],dfJPInew2['B'])
modelI.fit(dfJPInew2[['Tahun']],dfJPInew2['I'])
```

```python
dfJPInew2['predJB'] = modelJB.coef_[0] * dfJPInew2['Tahun'] + modelJB.intercept_
dfJPInew2['predB'] = modelB.coef_[0] * dfJPInew2['Tahun'] + modelB.intercept_
dfJPInew2['predI'] = modelI.coef_[0] * dfJPInew2['Tahun'] + modelI.intercept_
dfJPInew2
```

```python
# plt.style.use('Seaborn')
plt.figure(figsize=(12,6))
plt.plot(dfJPInew2['Tahun'],dfJPInew2['JB'],'g-o')
plt.plot(dfJPInew2['Tahun'],dfJPInew2['B'],'b-o')
plt.plot(dfJPInew2['Tahun'],dfJPInew2['I'],'r-o')

plt.plot(dfJPInew2['Tahun'],dfJPInew2['predJB'],'y-')
plt.plot(dfJPInew2['Tahun'],dfJPInew2['predB'],'y-')
plt.plot(dfJPInew2['Tahun'],dfJPInew2['predI'],'y-')


plt.xlabel('Tahun'); plt.ylabel('jumlah penduduk(ratus juta jiwa)')
plt.yticks(np.arange(0,300000000,50000000))
plt.legend([dfJPI['Provinsi'].iloc[11],dfJPI['Provinsi'].iloc[6],dfJPI['Provinsi'].iloc[33],'Best fit Line'])
plt.show()
```

```python
print(f'Prediksi jumlah penduduk {dfJPI["Provinsi"].iloc[11]} di tahun 2050 : ' + str(int(modelJB.predict([[2050]])[0])))
print(f'Prediksi jumlah penduduk {dfJPI["Provinsi"].iloc[6]} di tahun 2050 : ' + str(int(modelB.predict([[2050]])[0])))
print(f'Prediksi jumlah penduduk {dfJPI["Provinsi"].iloc[33]} di tahun 2050 : ' + str(int(modelI.predict([[2050]])[0])))
```

## Soal 2

```python
dfFifa = pd.read_csv('data.csv')
dfFifa = dfFifa.drop(columns = 'Unnamed: 0')
dfFifa.head()
```

```python
def target(row):
    if row['Age'] <= 25 and row['Overall'] >= 80 and row['Potential']>=80 :
        return 'Target'
    else :
        return 'Non-Target'
dfFifa['hasil'] = dfFifa.apply(target, axis=1)
dfFifa['hasilangka'] = dfFifa['hasil'].apply(lambda x: 1 if x == 'Target' else 0)
dfFifa.head()
```

<hr>

### 2.1

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.plot(dfFifa['Age'][dfFifa['hasil']=='Target'],dfFifa['Overall'][dfFifa['hasil']=='Target'],'g.')
plt.plot(dfFifa['Age'][dfFifa['hasil']=='Non-Target'],dfFifa['Overall'][dfFifa['hasil']=='Non-Target'],'r.')
plt.title('Age vs Overall'); plt.xlabel('Age'); plt.ylabel('Overall')
plt.legend(['Target','Non-Target'])

plt.subplot(122)
plt.plot(dfFifa['Age'][dfFifa['hasil']=='Target'],dfFifa['Potential'][dfFifa['hasil']=='Target'],'g.')
plt.plot(dfFifa['Age'][dfFifa['hasil']=='Non-Target'],dfFifa['Potential'][dfFifa['hasil']=='Non-Target'],'r.')
plt.title('Age vs Potential'); plt.xlabel('Age'); plt.ylabel('Potential')
plt.legend(['Target','Non-Target'])
```

<hr>

### 2.2

```python
from sklearn.model_selection import train_test_split

xtr, xts, ytr, yts = train_test_split(dfFifa[['Age','Overall','Potential']],dfFifa['hasilangka'], test_size = 0.2)
dtrain = pd.concat([xtr,ytr], axis=1)
dtes = pd.concat([xts,yts], axis=1)
```

```python
from sklearn.model_selection import KFold
k = KFold(n_splits = 10)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_graphviz
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
```

```python
score1 = []
score2 = []
score3 = []

for train,validate in k.split(dtrain):
    xtrain = dtrain.iloc[train][['Age','Overall','Potential']]
    ytrain = dtrain.iloc[train]['hasilangka']
    xvalid = dtrain.iloc[validate][['Age','Overall','Potential']]
    yvalid = dtrain.iloc[validate]['hasilangka']
    
    # model1
    model1 = LogisticRegression()
    model1.fit(xtrain,ytrain)
    score1.append(model1.score(xvalid,yvalid))
    
    # model2
    model2 = DecisionTreeClassifier(criterion='gini')
    model2.fit(xtrain,ytrain)
    score2.append(model2.score(xvalid,yvalid))
    
    # model3
    model3 = LinearRegression()
    model3.fit(xtrain,ytrain)
    score3.append(model3.score(xvalid,yvalid))
```

```python
print(f'Score Model Logistic Regression = {np.mean(score1)} ')
print(f'Score Model Decision Tree = {np.mean(score2)} ')
print(f'Score Model Linear Regression= {np.mean(score3)} ')
```
