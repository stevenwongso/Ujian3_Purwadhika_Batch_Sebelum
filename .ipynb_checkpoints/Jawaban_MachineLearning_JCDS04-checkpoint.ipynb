```python
from sklearn.preprocessing import LabelEncoder
```

```python
def alcohol(x) :
    if x == 'several times a day' :
        return 0
    elif x == 'every day' :
        return 1
    elif x == 'several times a week' :
        return 2
    elif x == 'once a week' :
        return 3
    else :
        return 4

def smoking(x):
    if x == 'never' :
        return -1
    elif x == 'occasional' :
        return 0
    else :
        return 1

def fever(x):
    if x == 'less than three months ago' :
        return -1
    elif x == 'more than three months ago' :
        return 0
    else :
        return 1
```

```python
dfDKnew=pd.DataFrame()
label = LabelEncoder()
dfDKnew['Age']=dfDK['Age']
dfDKnew['Childish diseases']=label.fit_transform(dfDK['Childish diseases'])
dfDKnew['Accident or serious trauma']=label.fit_transform(dfDK['Accident or serious trauma'])
dfDKnew['Surgical intervention']=label.fit_transform(dfDK['Surgical intervention'])
dfDKnew['High fevers in the last year']=dfDK['High fevers in the last year'].apply(fever)
dfDKnew['Frequency of alcohol consumption']=dfDK['Frequency of alcohol consumption'].apply(alcohol)
dfDKnew['Smoking habit']=dfDK['Smoking habit'].apply(smoking)
dfDKnew['Number of hours spent sitting per day']=dfDK['Number of hours spent sitting per day']
dfDKnew['Diagnosis']=dfDK['Diagnosis']
```

```python
dfDKnew
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_graphviz
from sklearn.neighbors import KNeighborsClassifier
```

```python
def k(x) :
    k = int(np.sqrt(x))
    if k % 2 == 0:
        k += 1
        return k
    else :
        return k
```

```python
# model1
model1 = LogisticRegression(solver='liblinear')
model1.fit(dfDKnew[['Age', 'Childish diseases', 'Accident or serious trauma',
       'Surgical intervention', 'High fevers in the last year',
       'Frequency of alcohol consumption', 'Smoking habit',
       'Number of hours spent sitting per day']],dfDK['Diagnosis']) 

# model2
model2 = DecisionTreeClassifier(criterion='gini')
model2.fit(dfDKnew[['Age', 'Childish diseases', 'Accident or serious trauma',
       'Surgical intervention', 'High fevers in the last year',
       'Frequency of alcohol consumption', 'Smoking habit',
       'Number of hours spent sitting per day']],dfDK['Diagnosis']) 
    
# model3
model3 = KNeighborsClassifier(n_neighbors= k(len(dfDK)))
model3.fit(dfDKnew[['Age', 'Childish diseases', 'Accident or serious trauma',
       'Surgical intervention', 'High fevers in the last year',
       'Frequency of alcohol consumption', 'Smoking habit',
       'Number of hours spent sitting per day']],dfDK['Diagnosis']) 
```

```python
dfPrediksinew=pd.DataFrame()
dfPrediksinew['Age']=dfPrediksi['Age']
dfPrediksinew['Childish diseases']=label.fit_transform(dfPrediksi['Childish diseases'])
dfPrediksinew['Accident or serious trauma']=label.fit_transform(dfPrediksi['Accident or serious trauma'])
dfPrediksinew['Surgical intervention']=label.fit_transform(dfPrediksi['Surgical intervention'])
dfPrediksinew['High fevers in the last year']=dfPrediksi['High fevers in the last year'].apply(fever)
dfPrediksinew['Frequency of alcohol consumption']=dfPrediksi['Frequency of alcohol consumption'].apply(alcohol)
dfPrediksinew['Smoking habit']=dfPrediksi['Smoking habit'].apply(smoking)
dfPrediksinew['Number of hours spent sitting per day']=dfPrediksi['Number of hours spent sitting per day']
```

```python
dfPrediksinew
```

```python
dfPrediksinew['Pred1'] = model1.predict(dfPrediksinew[['Age', 'Childish diseases', 'Accident or serious trauma',
       'Surgical intervention', 'High fevers in the last year',
       'Frequency of alcohol consumption', 'Smoking habit',
       'Number of hours spent sitting per day']])
dfPrediksinew['Pred2'] = model2.predict(dfPrediksinew[['Age', 'Childish diseases', 'Accident or serious trauma',
       'Surgical intervention', 'High fevers in the last year',
       'Frequency of alcohol consumption', 'Smoking habit',
       'Number of hours spent sitting per day']])
dfPrediksinew['Pred3'] = model3.predict(dfPrediksinew[['Age', 'Childish diseases', 'Accident or serious trauma',
       'Surgical intervention', 'High fevers in the last year',
       'Frequency of alcohol consumption', 'Smoking habit',
       'Number of hours spent sitting per day']])
```

```python
dfbook = pd.read_csv("books.csv")
dfrating = pd.read_csv("ratings.csv")
```

```python
display(dfbook.columns, dfrating.columns)
display(dfbook.head(1))
display(dfrating.head(1))
```

```python
dfbook['feature']= dfbook.apply(lambda x : str(x['authors']) + ' ' + str(x['original_title']) + ' ' + str(x['title']) + ' '  + str(x['language_code']),axis=1)

dfbook['feature'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
cm = cv.fit_transform(dfbook['feature'])
cosScore = cosine_similarity(cm.toarray())
```

```python
andi1 = dfbook[dfbook['original_title']=='The Hunger Games']['book_id'].index.values[0]
andi2 = dfbook[dfbook['original_title']=='Catching Fire']['book_id'].index.values[0] 
andi3 = dfbook[dfbook['original_title']=='Mockingjay']['book_id'].index.values[0]
andi4 = dfbook[dfbook['original_title']=='The Hobbit or There and Back Again']['book_id'].index.values[0] 
suka1 = [andi1,andi2,andi3,andi4]

similarbook0 = list(enumerate(cosScore[suka1[0]]))
similarbook1 = list(enumerate(cosScore[suka1[1]]))
similarbook2 = list(enumerate(cosScore[suka1[2]]))
similarbook3 = list(enumerate(cosScore[suka1[3]]))

similarbook=[]
for i in range(len(similarbook0)):
    new = (similarbook0[i][0], (similarbook0[i][1]+similarbook1[i][1]+similarbook2[i][1]+similarbook3[i][1])/4)
    similarbook.append(new)

similarbook = sorted(similarbook, key=lambda x: x[1], reverse=True)    

j = 0
print('Rekomendasi buku untuk Andi : ')
for i in similarbook:
    if j < 5 and i[0] not in suka1:
        print('     - ' + str(dfbook['title'].iloc[i[0]]))
        j += 1
    elif j >= 5 :
        break
    else :
        pass
```

```python
dfPokemon = pd.read_csv('pokemon.csv')
dfPokemon.head()
```

```python
dfCombat = pd.read_csv('combats.csv')
dfCombat.head()
```

```python
dfCombat['WinnerNumber']= dfCombat.apply(lambda x: 1 if x['Winner'] == x['First_pokemon'] else 2, axis = 1)
```

```python
dffeature = dfCombat.merge(dfPokemon, left_on='First_pokemon', right_on='#').merge(dfPokemon, left_on='Second_pokemon', right_on='#',suffixes=('_1', '_2'))
dffeature = dffeature.drop(columns=['First_pokemon','Second_pokemon','Winner'])
dffeature = dffeature.drop(columns=['#_1','Name_1','Type 1_1','Type 2_1','Legendary_1','Generation_1'])
dffeature = dffeature.drop(columns=['#_2','Name_2','Type 1_2','Type 2_2','Legendary_2','Generation_2'])
dffeaturex = dffeature.drop(columns=['WinnerNumber']).copy()
dffeaturey = dffeature['WinnerNumber'].copy()
dffeaturex.head()
```

```python
from sklearn.model_selection import train_test_split

xtr, xts, ytr, yts = train_test_split(dffeaturex,dffeaturey, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
modelcombat= LogisticRegression()
modelcombat.fit(xtr,ytr)
```

```python
pred = modelcombat.predict(xts)
```

```python
modelcombat.predict_proba(xts)
```
